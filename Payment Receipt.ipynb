Importing libraries


```
from reportlab.lib.pagesizes import letter
from reportlab.pdfgen import canvas
```

Createing a canvas

```
def create_receipt():
    # Create a canvas
    c = canvas.Canvas("receipt.pdf", pagesize=letter)
    width, height = letter

    # Draw the receipt contents
    c.setFont("Helvetica", 12)

    # Shop Name and Address
    c.drawCentredString(width / 2.0, height - 40, "SHOP ABC")
    c.setFont("Helvetica", 10)
    c.drawCentredString(width / 2.0, height - 60, "Address: Lorem Ipsum, 23-10")
    c.drawCentredString(width / 2.0, height - 75, "Telp. 0123456789")

    # Horizontal Line
    c.line(30, height - 90, width - 30, height - 90)

    # Receipt title
    c.setFont("Helvetica", 12)
    c.drawCentredString(width / 2.0, height - 110, "PAYMENT RECEIPT")

    # Horizontal Line
    c.line(30, height - 125, width - 30, height - 125)

    # Description and Price Headers
    c.setFont("Helvetica-Bold", 10)
    c.drawString(40, height - 140, "Description")
    c.drawString(400, height - 140, "Price")

    # Items
    items = [("Dress shirt", "1,500"), ("Chino pants", "2,000"), ("Denim jeans", "2,500"), ("Cotton t-shirt", "500")]
    y = height - 160
    c.setFont("Helvetica", 10)
    for item, price in items:
        c.drawString(40, y, item)
        c.drawString(400, y, "Rs " + price)
        y -= 20

    # Total
    c.drawString(41, y - 1, "---------------------------------------------------------------------------------------------------------------------------------")
    c.drawString(40, y - 10, "Total")
    c.drawString(400, y - 10, "Rs 6,500")

    # Payment details
    c.drawString(40, y - 30, "Cash")
    c.drawString(400, y - 30, "Rs 6,500")
    c.drawString(40, y - 50, "Change")
    c.drawString(400, y - 50, "Rs 0")

    # Thank you note
    c.drawCentredString(width / 2.0, y - 100, "Thank you for your payment!")
    c.drawCentredString(width / 2.0, y - 118, "Please keep this receipt for your records.")
    c.save()
```

```
create_receipt()
```
